```python
from dqt import DBUtils, get_fut_min, get_fut_taq
import pandas as pd
```

```python
dbu = DBUtils()
varieties = ['SP', 'EG']#CY，SN，FU，'SC'
varieties = pd.Series(varieties)

all_varieties = ['A', 'AG', 'AL', 'AU', 'AP', 'BU', 'B', 'C', 'CF', 'CJ', 'CS', 'CU', 'CY', 'EG', 'FG', 'HC', 'I', 'J',
                'JD', 'JM', 'L', 'M', 'MA', 'NI', 'NR', 'OI', 'P', 'PB', 'PP', 'RB', 'RM', 'RU', 'SC', 'SP', 'SR', 'T',
                'TA', 'TF', 'TS', 'UR', 'V', 'Y', 'ZC', 'ZN']

```

```python
c = list ()
for i,variety in varieties.iteritems():
    print('variety: ', variety)
    df = dbu.get_main_daily(variety=[variety], start_date='2014-01-01')
    summary_df = df.groupby('symbol').agg({'date': ['first', 'last']})
    print('summary_df: ', summary_df)
    sleep(1000)
    a = list()
    b = list()
    for i, contract in summary_df.iterrows():
        min_df = get_fut_min(contract.name, str(contract[('date', 'first')].date()), str(contract[('date', 'last')].date()))
        taq_df = get_fut_taq(contract.name, str(contract[('date', 'first')].date()), str(contract[('date', 'last')].date()))
        print(taq_df.groupby(taq_df.index))
        a.append(min_df.groupby('trading_date').count()['datetime'])
        b.append(taq_df.groupby(taq_df.index).count()['datetime'])
        del min_df
        del taq_df

    merge_a = pd.concat(a)
    merge_b = pd.concat(b)
    del a  
    del b

    c.append({'期货':variety,'taq数据少':merge_b[merge_b == merge_b.describe()['min']], '日数据存在而taq数据缺少':pd.Index(df['date']).difference(merge_b.index),'大于30000条记录':merge_b[merge_b>30000]})

c = pd.DataFrame(c)    
```

```python
print(c)
```

```python
c.to_csv("futures.csv")
```
