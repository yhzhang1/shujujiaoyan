```python
from dqt import DBUtils, get_fut_min, get_fut_taq
import pandas as pd
from time import sleep
import numpy as np
from time import time
from datetime import timedelta

```

```python
dbu = DBUtils()

all_varieties = ['A', 'AG', 'AL', 'AU', 'AP', 'BU', 'B', 'C', 'CF', 'CJ', 'CS', 'CU', 'CY', 'EG', 'FG', 'HC', 'I', 'J',
                'JD', 'JM', 'L', 'M', 'MA', 'NI', 'NR', 'OI', 'P', 'PB', 'PP', 'RB', 'RM', 'RU', 'SC', 'SP', 'SR', 'T',
                'TA', 'TF', 'TS', 'UR', 'V', 'Y', 'ZC', 'ZN']

all_trading_days = dbu.get_trading_day()
first_trading_day = all_trading_days[0]
print('first_trading_day: {}'.format(first_trading_day))
tick_data_first_day = '2014-01-01'
```

```python
all_trading_days
```

```python
#taq
main_daily_df = dbu.get_main_daily(variety=all_varieties, start_date=tick_data_first_day)
print('main_daily_df: \n', main_daily_df)
num_records_df = pd.DataFrame(data={'date':main_daily_df['date'], 'symbol':main_daily_df['symbol']}, columns=['date', 'symbol', 'num taq records'])
print('num_records_df: \n', num_records_df)

df_size = num_records_df.shape[0]
start_time = time()
i=0
for idx, row in main_daily_df.iterrows():
    i+=1
    if i%1000 == 0:
        end_time = time()
        duration = end_time - start_time
        all_duration = duration * (df_size / 1000)
        start_time = time()
        print('time approximation: {}'.format(timedelta(seconds=all_duration)))
        print('idx: ', idx)
        print('row[date]: ', row['date'])
        print('row[symbol]: ', row['symbol'])
    try:
        taq_df = get_fut_taq(row['symbol'], start_date=row['date'], end_date=row['date'])
    except:
        print('{} at {} has error get_fut_taq'.format(row['symbol'], row['date']))
        continue
    #print('taq_df: \n', taq_df)
    num_taq_records = taq_df.shape[0]
    del taq_df

    
    #print('num_records_df.loc[idx, num taq records]: ', num_records_df.loc[idx, 'num taq records'])
    num_records_df.loc[idx, 'num taq records'] = num_taq_records
    #print('num_records_df.loc[idx, num taq records]: ', num_records_df.loc[idx, 'num taq records'])
    

print('#####finished checking all days#####')
    
print('num_records_df: \n', num_records_df)

    

    





    

    
```

```python
num_records_df = num_records_df.dropna()
```

```python
num_records_df.to_csv('num_records_df.csv')
```

```python
one_pct_quantile = num_records_df['num taq records'].quantile(q=0.01)
print('one_pct_quantile: ', one_pct_quantile)
ninety_nine_pct_quantile = num_records_df['num taq records'].quantile(q=0.99)
print('ninety_nine_pct_quantile: ', ninety_nine_pct_quantile)
less_records_df = num_records_df[num_records_df['num taq records'] < one_pct_quantile]
less_records_df = less_records_df.sort_values(by='num taq records')
more_records_df = num_records_df[num_records_df['num taq records'] > ninety_nine_pct_quantile]
more_records_df = more_records_df.sort_values(by='num taq records', ascending=False)
```

```python
less_than_one_pct_TAQ = num_records_df[num_records_df['num taq records'] <= one_pct_quantile].sort_values(by='num taq records')
```

```python
less_than_one_pct_TAQ.to_csv('less_than_one_pct_TAQ.csv')
```

```python
#taq
duplicated_records_df = pd.DataFrame(columns=['date','symbol', 'num taq records'])
df_size = num_records_df.shape[0]
start_time = time()
i=0
for idx, row in num_records_df.iterrows():
    i+=1

    fut_taq = get_fut_taq(row['symbol'], start_date=row['date'], end_date=row['date'])
    #print('fut_taq: ', fut_taq)
    #has_duplicated_records = (fut_taq.groupby('datetime').count() >= 2).any(axis=None)
    has_duplicated_records = fut_taq.duplicated().any()
    if i % 1000 == 0:
        end_time = time()
        duration = end_time - start_time
        all_duration = duration * (df_size / 1000)
        start_time = time()
        print('time approximation: {}'.format(timedelta(seconds=all_duration)))
        print('i: ', i)
        print('idx: ', idx)
        #print('row: \n', row)
        print('row[symbol]: ', row['symbol'])
        print('row[date]: ', row['date'])
        print('type(row[date]): ', type(row['date']))    
        print('has_duplicated_records: ', has_duplicated_records)
    if has_duplicated_records:
        duplicated_records_df = duplicated_records_df.append({'date':row['date'], 'symbol':row['symbol'], 'num taq records':row['num taq records']}, ignore_index = True)
        
        
print('####finished checking all duplicated records##########')
    
```
